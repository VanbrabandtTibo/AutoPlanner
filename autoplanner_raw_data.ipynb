```python
from pulp import *
from domain.entities import DayOfTheWeek, Employee, Patient
```

```python
with open('data/therapists.json') as f:
   t = json.load(f)

with open('data/patients.json') as f:
   p = json.load(f)
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
therapists = [i for i in t.keys()]
patients = [i for i in p.keys()]
print(therapists)
print(patients)

nr_therapists = len(therapists)
for i in range(0,nr_therapists):
   employee = Employee(i)
```

```python
ct = list(itertools.combinations((therapists), 2 ))
ct.extend(therapists)
print(ct)
```

```python
ct_check = []
# check if the combination of therapists are available on the same day add the combination to ct_check
for therapist_combo in ct:
    if type(therapist_combo) == tuple:
        for day in days:
            if t[therapist_combo[0]]['Week'][day] == t[therapist_combo[1]]['Week'][day] is not None:
                if therapist_combo not in ct_check:
                    ct_check.append((therapist_combo, day))
    else:
        ct_check.append(therapist_combo)
print(ct_check)
```

```python
cp = list(itertools.combinations((patients), 8 ))
cp.extend(itertools.combinations((patients), 7 ))
cp.extend(itertools.combinations((patients), 6 ))
cp.extend(itertools.combinations((patients), 5 ))
cp.extend(itertools.combinations((patients), 4 ))
cp.extend(itertools.combinations((patients), 3 ))
cp.extend(itertools.combinations((patients), 2 ))
cp.extend(patients)
print(cp)
```

```python
cp_check = []
a = []
# check if the combination of patients are available on the same day add the combination to cp_check
for patient_combo in cp:
    if type(patient_combo) == tuple:
        for day in days:
            for patient in patient_combo:
                if p[patient]['Week'][day] is not None:
                    s,e = p[patient]['Week'][day]
                    a.append(((s,e),day))

            print(a)
            f = a[0]
            chk = True
            for item in a:
                if f != item:
                    chk = False
                    break
            if chk == True:
                cp_check.append((patient_combo, day))
            else:
                continue
    else:
        cp_check.append(patient_combo)
#print(cp_check)
#print(p_se)
```

```python
# Decision variables for ct and cp
ct_vars = LpVariable.dicts("ct", ct_check, 0, 1, LpInteger)
cp_vars = LpVariable.dicts("cp", cp, 0, 1, LpInteger)
print(ct_vars)
print(cp_vars)
```

```python
# Create a maximization problem
prob = LpProblem("Automated Scheduler", LpMaximize)
print(prob)
```

```python
# Objective function
prob += lpSum([ct_vars[i] for i in ct]) + lpSum([cp_vars[i] for i in cp])
```

```python
print(p['Patient 1']['Week']['Monday'])
```

```python
# Constraint to check if the combinations of patients matches the same timeslot
for i in patients:
    for j in days:
        prob += lpSum([cp_vars[k] for k in cp if i in k]) <= 1
```

```python
# Solve the problem
status = prob.solve()
print("Status:", LpStatus[status])
```

```python
# Print schedule for each patient and therapist
for i in patients:
    for j in days:
        for k in cp:
            if i in k and cp_vars[k].value() == 1:
                print(i, j, k)
```
